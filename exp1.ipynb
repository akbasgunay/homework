```python
import numpy as np
import matplotlib.pyplot as plt

def best_fit_line(x, y):
    m, b = np.polyfit(x, y, 1) # Slope and intercept of lines
    best_fit = m*x + b
    plt.plot(x, y, 'o') # Creating a scatter plot
    plt.plot(x, best_fit) # Best fit line        

    return m,b
```

```python
# Revolution values measured in a minute
x = np.array([3.5, 4, 4.5, 5, 6, 8, 10]) # Input voltage
y = np.array([770, 910, 1028, 1156, 1418, 1942, 2496]) # Corresponding revolution value measured in one minute

best_fit_line(x, y)

plt.xlabel('Volt') 
plt.ylabel('Rpm') 
plt.title("Line of Best Fit")
```
