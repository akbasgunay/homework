# Experiment 1 Slotted Opto Transducer

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.metrics import mean_squared_error
import math

def best_fit_line(x, y):
    m, b = np.polyfit(x, y, 1) # Slope and intercept of lines
    best_fit = m*x + b
    plt.plot(x, y, 'o') # Creating a scatter plot
    plt.plot(x, best_fit) # Best fit line        

    return m,b
```

```python
# Revolution values measured in a minute
x = np.array([3.5, 4, 4.5, 5, 6, 8, 10]) # Input voltage
y = np.array([770, 910, 1028, 1156, 1418, 1942, 2496]) # Corresponding revolution value measured in one minute

best_fit_line(x, y)

plt.xlabel('Volt') 
plt.ylabel('Rpm') 
plt.title("Line of Best Fit")
```

```python
# Revolution values measured in a second then multiplied by 60
x = np.array([3.5, 4, 4.5, 5, 6, 8, 10]) # Input voltage
y = np.array([780, 900, 1080, 1140, 1440, 1980, 2520]) # Corresponding rev per sec times 60

best_fit_line(x, y)

m, b = best_fit_line(x,y)
predict_y = 1800 # 30 rev per sec multiplied w/60
predict_x = (predict_y - b) / m
print("Corresponding voltage to 30 rev per sec: ", predict_x)

plt.xlabel('Volt') 
plt.ylabel('Rpm') 
plt.title("Line of Best Fit")
```

```python
relative_percent_error = (abs(7.44 - predict_x) / predict_x) * 100

print("The relative error is", relative_percent_error, "%")
```

# Experiment 3 Hall Effect Transducer && Slotted Opto Transducer

```python
# Revolution values measured in a second using Hall effect transducer
x = np.array([3.5, 5, 7, 10]) # Input voltage
y = np.array([13, 19, 28, 42]) # Corresponding rev per sec

best_fit_line(x, y)
plt.xlabel('Volt') 
plt.ylabel('Rev per sec') 
plt.title("Line of Best Fit")
```

```python
# Revolution values measured in a second using slotted opto transducer
x = np.array([3.5, 5, 7, 10]) # Input voltage
y = np.array([13, 19, 29, 42]) # Corresponding rev per sec

best_fit_line(x, y)
plt.xlabel('Volt') 
plt.ylabel('Rev per sec') 
plt.title("Line of Best Fit")
```

```python
rps_hall = [13, 19, 28, 42] # Revolution per second using Hall effect transducer
rps_slotted = [13, 19, 29, 42] # Revolution per secod using Slotted effect transducer

mse = sklearn.metrics.mean_squared_error(rps_hall, rps_slotted) # Mean square error

rmse = math.sqrt(mse) # Root mean squre error between these two outputs

print("Root mean squared error is", rmse)
```
